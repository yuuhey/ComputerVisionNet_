```python
### Model
resnet18 = models.resnet18(pretrained=True)

## Dataset
to_tensor = torchvision.transforms.Compose(
                [torchvision.transforms.ToTensor(),
               torchvision.transforms.Normalize(mean=[0.485, 0.456, 0.406],
                                 std=[0.229, 0.224, 0.225])]
                                          )

cifar10 = torchvision.datasets.CIFAR10(root='./', download=True, transform=to_tensor)

dataloader = torch.utils.data.DataLoader(cifar10, batch_size=8, shuffle=True, num_workers=2)
```

```python
for idx, data in enumerate(dataloader):
    
    img, gt = data
    
    print(img.shape)
    
    scores = resnet18(img)
    
    print(scores.shape)
    break
```

## Fine-tuning

```python
## TODO: 1. replace the last FC layer for cifar10
### Hint: 1000 -> 10
num_classes = 10

resnet18.fc = torch.nn.Linear(in_features=resnet18.fc.in_features, out_features=num_classes)

# 학습 파라미터 초기화 : 마지막 FC레이어를 변경했으므로 해당 레어어의 가중치들을 새로운 랜덤 초기값으로 설정
def weights_init(m):
    if isinstance(m, torch.nn.Linear):
        torch.nn.init.xavier_uniform_(m.weight)
        if m.bias is not None:
            torch.nn.init.zeros_(m.bias)

resnet18.fc.apply(weights_init)
# 주의할 점은, 미리 학습된 모델의 가중치를 초기화하면 이전에 학습한 정보를 잃게 되므로 새로운 학습을 수행해야 합니다.
```
